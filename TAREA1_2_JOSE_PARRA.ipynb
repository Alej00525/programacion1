# Pasar de **lista** a tupla, diccionario y conjunto.

```python
Llista= ['p','a','r','r','a',0,5,2,5]
print(Llista)
Ltupla = tuple(Llista)
print(Ltupla)
Ldiccionario = dict(enumerate(Llista))
print(Ldiccionario)
Lconjunto = set(Llista)
print(Lconjunto)
```

# Pasar de **tupla** a lista, diccionario y conjunto.

```python
Ttupla= ('p','a','r','r','a',0,5,2,5)
print(Ttupla)
Tlista = list(Ttupla)
print(Tlista)
Tdiccionario = dict(enumerate(Ttupla))
print(Tdiccionario)
Tconjunto = set(Ttupla)
print(Tconjunto)
```

## Pasar de **diccionario** a lista, tupla y conjunto.

```python
Ddiccionario = {'Apellido':'Parra','Cumpleaños':'Mayo 25'}
print(Ddiccionario)
Dlista = list(Ddiccionario.items())
print(Dlista)
Dtupla = tuple(Ddiccionario.items())
print(Dtupla)
Dconjunto = set(Ddiccionario.items())
print(Dconjunto)
# aqui la lista, tupla y conjunto se hicieron a partir de las tuplas CLAVE:VALOR
```

```python
Ddiccionario = {'Apellido':'Parra','Cumpleaños':'Mayo 25'}
print(Ddiccionario)
Dlista = list(Ddiccionario)+list(Ddiccionario.values())
print(Dlista)
Dtupla = tuple(Dlista)
print(Dtupla)
Dconjunto = set(Dlista)
print(Dconjunto)
# aqui se hizo a partir de la lista de clave + valor para que dentro de la lista, tupla y conjunto las claver y valores quedaran mezcladas
```

## Pasar de **conjunto** a lista, tupla y diccionario.

```python
Cconjunto = {'p','a','r','r','a',0,5,2,5}
print(Cconjunto) # aqui se ve como elimina los duplicados
Clista = list(Cconjunto)
print(Clista)
Ctupla = tuple(Cconjunto)
print(Ctupla)
Cdiccionario = dict(enumerate(Cconjunto))
print(Cdiccionario)
```
